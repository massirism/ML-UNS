```python
#Importo bibliotecas
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
from sklearn import metrics
```

```python
#Numero de muestras
muestras=50

#Media y desvio del primer conjuto de datos
media1 = 20 
desvio1 = 5

#Media y desvio del segundo conjuto de datos
media2 = 30 
desvio2 = 3

#Genero las 2 distribuciones Gaussianas
A = random.normal(media1,desvio1,muestras)
B = random.normal(media2,desvio2,muestras)


#print("B =",B)
```

```python
sns.distplot(A,hist=True)
sns.distplot(B,hist=True)
#Lujo.
sns.set_style('darkgrid')
#Muestro en pantalla
plt.show()
```

```python
Av=[0]*len(A) #Variable del tipo "A"
Bv=[1]*len(B) #Variable del tipo "B"


A_labeledData=np.column_stack((Av,A))  #Concateno los valores de "A" con su etiqueta.

B_labeledData=np.column_stack((Bv,B))  #Concateno los valores de "B" con su etiqueta.

#Ahora uno los dos conjuntos de valores.
DatosEtiquetados=np.concatenate((A_labeledData,B_labeledData),axis=0)


#Creo el df
df=pd.DataFrame(DatosEtiquetados,columns=["Tipo de dato","Dato"])

#Paso todos los valores de la columna Dato de str a float.
df.loc[:,"Dato"] = pd.to_numeric(df.loc[:,"Dato"])

#Ahora ordeno el dataframe en forma ascendente
df=df.sort_values("Dato")
df=df.reset_index(drop=True)

print(df)
```

```python
#Grafico la curva ROC.

fpr, tpr, thresholds = metrics.roc_curve(df["Tipo de dato"], df["Dato"])

roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc,estimator_name='example estimator')
display.plot()  
plt.show()      
```
